```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
train_df=pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/train.csv')
```

```python
train_df.columns
```

```python
num_df=train_df.select_dtypes(include='number')
correlation_matrix = num_df.corr()
correlation_matrix['SalePrice'].sort_values(ascending = False)
```

```python
r_t = ["GarageArea","OverallQual","TotalBsmtSF","1stFlrSF","2ndFlrSF","LowQualFinSF","GrLivArea","BsmtFullBath","BsmtHalfBath","FullBath","HalfBath","TotRmsAbvGrd","SalePrice"]
```

```python
se_t=train_df[r_t]
```

```python
se_t = se_t.assign(
    TotalBath=lambda df: (df['BsmtFullBath'].fillna(0) + 
                           df['BsmtHalfBath'].fillna(0) + 
                           df['FullBath'].fillna(0) + 
                           df['HalfBath'].fillna(0)),
    
    TotalSF=lambda df: (df['TotalBsmtSF'].fillna(0) + 
                        df['1stFlrSF'].fillna(0) + 
                        df['2ndFlrSF'].fillna(0) + 
                        df['LowQualFinSF'].fillna(0) + 
                        df['GrLivArea'].fillna(0))
)
```

```python
se_t
```

```python
train_df = se_t[['TotRmsAbvGrd','TotalBath','GarageArea','TotalSF','OverallQual','SalePrice']]
```

```python
train_df
```

```python
from sklearn.model_selection import train_test_split
train_set,test_set=train_test_split(train_df,test_size=0.2,random_state=42)
print(f" Rows in Train_set : {len(train_set)}\n Rows in test set: {len(test_set)}")
```

```python
hsing = train_set.drop("SalePrice",axis=1)
hsing_label=train_set['SalePrice'].copy()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('std_scaler', StandardScaler())
])
```

```python
x_train=my_pipeline.fit_transform(hsing)
```

```python
x_train
```

```python
y_train=hsing_label
```

```python
y_train
y_train.shape
```

```python
test_df=pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/test.csv')
```

```python
req_tst = ["GarageArea","OverallQual","TotalBsmtSF","1stFlrSF","2ndFlrSF","LowQualFinSF","GrLivArea","BsmtFullBath","BsmtHalfBath","FullBath","HalfBath","TotRmsAbvGrd"]
```

```python
se_tst=test_df[req_tst]
```

```python
se_tst = se_tst.assign(
    TotalBath=lambda df: (df['BsmtFullBath'].fillna(0) + 
                           df['BsmtHalfBath'].fillna(0) + 
                           df['FullBath'].fillna(0) + 
                           df['HalfBath'].fillna(0)),
    
    TotalSF=lambda df: (df['TotalBsmtSF'].fillna(0) + 
                        df['1stFlrSF'].fillna(0) + 
                        df['2ndFlrSF'].fillna(0) + 
                        df['LowQualFinSF'].fillna(0) + 
                        df['GrLivArea'].fillna(0))
)
```

```python
se_tst
```

```python
test_df_unproc = se_tst[['TotRmsAbvGrd','TotalBath','GarageArea','TotalSF','OverallQual']]
```

```python
test_df_unproc
```

```python
test_df = test_df_unproc.fillna(test_df_unproc.mean())
```

```python
x_test = my_pipeline.transform(test_df[['TotRmsAbvGrd','TotalBath','GarageArea','TotalSF','OverallQual']].values)
```

```python
x_test
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_train_pred=model.predict(x_train)
```

```python
train_mse=mean_squared_error(y_train,y_train_pred)
```

```python
train_rmse=np.sqrt(train_mse)
```

```python
print(f"Training MSE: {train_mse:.2f}, Training RMSE: {train_rmse:.2f}")
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,x_train,y_train,scoring="neg_mean_squared_error",cv = 200)
rmse_scores = np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(scores):
    print("Scores:",scores)
    print("Mean:",scores.mean())
    print("Standard Deviation",scores.std())
```

```python
print_scores(rmse_scores)
```
